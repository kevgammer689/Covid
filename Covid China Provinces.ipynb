```python
import pandas as pd

# we are using the inline backend
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv")
```

```python
provincias = data[data["Country/Region"]=="China"]

provincias.head()
```

```python
provincias.set_index('Province/State', inplace=True)

provincias.head()
```

```python
provincias = provincias.drop(["Country/Region","Lat","Long"],axis=1)

provincias.head()
```

```python
provincias.columns = list(map(str, provincias.columns))
[print (type(x)) for x in provincias.columns.values] #<-- uncomment to check type of column headers
```

```python
provincias = provincias.transpose()

provincias.columns
```

```python
provincias
```

##  Visualization

```python
Hubei = provincias.loc[:,['Hubei']]

```

```python
Hubei.plot(kind='line',figsize=(12, 8))

plt.title('Confirmed Hubei')
plt.ylabel('Number of Positives')
plt.xlabel('Days')

plt.show() # need this line to show the updates made to the figure
```

```python
top_eight = provincias.loc[:,['Guangdong', 'Henan', 'Zhejiang', 'Hunan', 'Anhui', 'Jiangxi','Shandong', 'Jiangsu']]
```

```python
#top_eight = top_5.index.map(int) # let's change the index values of df_top5 to type integer for plotting
top_eight.plot(kind='line', figsize=(14, 8)) # pass a tuple (x, y) size


plt.title('Confirmed Cases in Provinces Out of Hubei')
plt.ylabel('Number of Confirmed Cases')
plt.xlabel('Days')

plt.show()
```

### Last 10 Days

```python
Hubei_last_ten = provincias.iloc[-20:-1,0]
```

```python
Hubei_last_ten.plot(kind='line',figsize=(12, 8))

plt.title('Confirmed Hubei Last 20 Days')
plt.ylabel('Number of Positives')
plt.xlabel('Days')

plt.show() # need this line to show the updates made to the figure
```

### Top 8 Provinces In Last Ten Days

```python
top_eight_last10 = provincias.iloc[-20:-1,1:9]
```

```python
top_eight_last10.plot(kind='line', figsize=(14, 8)) # pass a tuple (x, y) size


plt.title('Top 8 Provinces Confirmed Cases in Last 20 Days')
plt.ylabel('Number of Confirmed Cases')
plt.xlabel('Days')

plt.show()
```

```python
top_10_20 = provincias.iloc[-20:-1,9:19]
```

```python
top_10_20.plot(kind='line', figsize=(14, 8)) # pass a tuple (x, y) size


plt.title('Top 10 to 20 Provinces Confirmed Cases in Last 20 Days')
plt.ylabel('Number of Confirmed Cases')
plt.xlabel('Days')
plt.legend(loc='best')

plt.show()
```

### CHINA

```python
provincias["total"] = provincias.iloc[:,0:].sum(axis=1)

provincias.head()
```

```python
China = provincias["total"]

China.head()
```

```python
China.plot(kind='line', figsize=(14, 8)) # pass a tuple (x, y) size


plt.title('Total China')
plt.ylabel('Number of Confirmed Cases')
plt.xlabel('Days')
plt.legend(loc='best')

plt.show();
```

### Last 20 Days

```python
china_last_20 = China.tail(20)

china_last_20.plot(kind='line', figsize=(14, 8)) # pass a tuple (x, y) size


plt.title('Total China Last 20 Days')
plt.ylabel('Number of Confirmed Cases')
plt.xlabel('Days')
plt.legend(loc='best')

plt.show();
```
